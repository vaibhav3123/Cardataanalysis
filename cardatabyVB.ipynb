```python
plt.figure(figsize=(15, 6))
plt.bar(car['Year'], car['Price'], color='skyblue')
plt.xlabel("Year")
plt.ylabel("Price")
plt.title("Bar Plot of Car Prices by Year")
plt.show()
```

```python
brandcount = car['Brand'].value_counts()
brandcount
```

```python
brandcount = car['Brand'].value_counts().plot(kind='pie',autopct='%1.1f%%')
plt.title('Brand Count')
plt.show()
```

```python
x = car.groupby('Brand')['Model'].unique()
x
```

```python
x = car.groupby('Brand')['Model'].value_counts()
x
```

```python
x.plot(kind='bar', stacked=True, figsize=(12, 8), colormap='viridis')
plt.title('Model Counts per Brand')
plt.xlabel('Brand')
plt.ylabel('Count of Models')
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
year = car.groupby('Brand')['Year'].value_counts().sort_index(ascending = False)
year
```

```python
plt.figure(figsize=(7, 8))
car['Year'].value_counts().plot(kind = 'pie',autopct='%1.1f%%')
plt.show()
```

```python
car.head(4)
```

```python
sns.countplot(x='Fuel_Type', data=car,palette='viridis')
plt.show()
for count in car['Fuel_Type'].value_counts():
    print(count)
```

```python
numeric_data = car.select_dtypes(include=["number"])
correlation_matrix = numeric_data.corr()
plt.figure(figsize=(10, 6))
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", fmt=".2f", linewidths=0.5)
plt.title("Feature Correlation Heatmap")
plt.show()
```

```python
sns.countplot(x='Transmission', data=car,palette='viridis')
plt.show()
for count in car['Transmission'].value_counts():
    print(count)
```

```python
plt.figure(figsize=(8, 5))
sns.boxplot(x=car["Owner_Count"], y=car["Price"], palette="coolwarm")

plt.xlabel("Number of Owners")
plt.ylabel("Car Price ($)")
plt.title("Impact of Owner Count on Car Price")
plt.show()

owner_price_comparison = car.groupby("Owner_Count")["Price"].mean().reset_index()
print(owner_price_comparison)
```
